```python
import cv2
import pytesseract
from PIL import Image
import os
import re

# Define paths
video_path = "2.mp4"  # Replace with your video file path
output_frames_dir = "frames"
os.makedirs(output_frames_dir, exist_ok=True)

# Function to extract frames from the video
def extract_frames(video_path, output_dir, interval=30):
    cap = cv2.VideoCapture(video_path)
    frame_count = 0
    saved_count = 0

    while cap.isOpened():
        success, frame = cap.read()
        if not success:
            break
        if frame_count % interval == 0:  # Save frame every 'interval' frames
            frame_filename = os.path.join(output_dir, f"frame_{saved_count:04d}.jpg")
            cv2.imwrite(frame_filename, frame)
            saved_count += 1
        frame_count += 1

    cap.release()
    return saved_count

# Function to extract text from an image
def extract_text(image_path):
    image = Image.open(image_path)
    return pytesseract.image_to_string(image)

# Function to filter phone numbers from text
def extract_phone_numbers(text):
    phone_pattern = r'\+?\d[\d\s().-]{7,}\d'
    return re.findall(phone_pattern, text)

# Main function to process video and extract numbers
def process_video(video_path, output_frames_dir):
    print("Extracting frames...")
    frame_count = extract_frames(video_path, output_frames_dir)
    print(f"Extracted {frame_count} frames.")

    all_numbers = set()
    for frame_file in sorted(os.listdir(output_frames_dir)):
        frame_path = os.path.join(output_frames_dir, frame_file)
        text = extract_text(frame_path)
        numbers = extract_phone_numbers(text)
        all_numbers.update(numbers)

    return sorted(all_numbers)

# Run the script
if __name__ == "__main__":
    print("Processing video for contact numbers...")
    phone_numbers = process_video(video_path, output_frames_dir)

    print("\nExtracted Contact Numbers:")
    for idx, number in enumerate(phone_numbers, start=1):
        print(f"{idx}. {number}")
```
